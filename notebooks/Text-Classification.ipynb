## Preprocessing
The text document is in csv format. It has 3 columns; questions, category 1, category 2. But it could not be loaded from pandas since the questions include comma and column separator is also comma. ### Preprocessing steps
    1. Auto correcting words
    2. Removing punctuations and numeric values
    3. Remove all the special characters
    4. Remove all single characters
    5. Remove single characters from the start
    6. Substituting multiple spaces with single space
    7. Removing prefixed 'b'
    8. Converting to Lowercase
    9. Lemmatization
    10. Stop word removal
    11. Replace new line character in questions 

```python
# Autocorrect words
spell = Speller(lang='en')
preprocessed_df["questions"] = preprocessed_df["questions"].apply(lambda x: spell(x))
```

### POS tags are generated after the second step since they are needed to be generated before loosing important words because of preprocessing.

```python
# Initial preprocessing to remove punctuations and numeric values for generating pos tags
remove_digits = str.maketrans('', '', digits)
translate_table = dict((ord(char), None) for char in string.punctuation)

def preprocess(content):
    return list(filter(None, content.translate(remove_digits).translate(translate_table).split(" ")))

# Generating POS tags
preprocessed_df["pos_tags"] = preprocessed_df["questions"].apply(lambda x : [(i[0].lower(),i[1]) for i in pos_tag(preprocess(x))])
```

```python
# Secondary preprocessing
X = preprocessed_df["questions"].values

documents = []

from nltk.stem import WordNetLemmatizer

stemmer = WordNetLemmatizer()

for sen in range(0, len(X)):
    # Remove all the special characters
    document = re.sub(r'\W', ' ', str(X[sen]))
    
    # Remove digits and punctutations
    document = " ".join(preprocess(document))

    # remove all single characters
    document = re.sub(r'\s+[a-zA-Z]\s+', ' ', document)
    
    # Remove single characters from the start
    document = re.sub(r'\^[a-zA-Z]\s+', ' ', document) 
    
    # Substituting multiple spaces with single space
    document = re.sub(r'\s+', ' ', document, flags=re.I)
    
    # Removing prefixed 'b'
    document = re.sub(r'^b\s+', '', document)
    
    # Converting to Lowercase
    document = document.lower()
    
    # Lemmatization
    document = document.split()

    document = [stemmer.lemmatize(word) for word in document]
    document = ' '.join(document)
    
    # Stop word removal
    document = [word for word in document.split() if word not in stopwords.words('english')]
    document = ' '.join(document)
    
    documents.append(document)

preprocessed_df["preprocessed_questions"] = documents
```

```python
# Replace new line character in categories
preprocessed_df["cat1"] = preprocessed_df["cat1"].apply(lambda x : x.replace("\n",""))
preprocessed_df["cat2"] = preprocessed_df["cat2"].apply(lambda x : x.replace("\n",""))
```

### 1. Bag of Words
The feature mentioned in the paper is text words + pos tags. So initially tokens were created as “word_pos-tag”. <br>
Eg:- word - “stay”     pos tag - “verb”     token -“stay_verb”<br>
But this caused to have a large sparse matrix. So only the words are taken for the feature.
CountVectorizer function in the Scikit-Learn Package is used in creating bag of words for each sentence.

```python
bow_vectorizer = CountVectorizer(max_features=100, min_df=5, max_df=0.7, stop_words=stopwords.words('english'))
bow_vector = bow_vectorizer.fit_transform(preprocessed_df["preprocessed_questions"].values).toarray()
```

### 2. Bigrams
Bi grams are also created using CountVectorizer function as bag of bi grams by setting ngram setting in the function.

```python
bigram_vectorizer = CountVectorizer(max_features=100,ngram_range=(2,2))
bigram_vector = bigram_vectorizer.fit_transform(preprocessed_df["preprocessed_questions"].values).toarray()
```

### 3. Head words
First Stanford core NLP parser was tried but it is not performance wise good. Then Malt parser is attempted. But installing it was quite hard due to some errors. Finally Spacy package is used to get head words. The words are tokenized and their tags are generated. Only the tags of "nsubj" and "nsubjpass" are identified as head words. The the head words are generated as bag of head words with CountVectorizer function with the use of custom head word tokenizer.

```python
# Head word tokenizer
nlp = spacy.load("en_core_web_sm")
def head_word_tokenizer(text):
    head_words = []
    for token in nlp(text):
        if token.dep_ == "nsubj" or token.dep_ == "nsubjpass":
            head_words.append(token.text)
            head_words.append(token.head.text)
    return head_words
```

```python
head_words_vectorizer = CountVectorizer(tokenizer = head_word_tokenizer,max_features=100,stop_words=stopwords.words('english'))
head_words_vector = head_words_vectorizer.fit_transform(preprocessed_df["questions"].values).toarray()
```

### 4. Head word Synonymns
Synonyms are retrieved from NLTK wordnet library. As earlier synonyms are generated as bag of head word synonyms with CountVectorizer function with the use of custom head word synonym tokenizer.

```python
def get_synonyms(word):
    syn = []
    for synset in wordnet.synsets(word):
       for lemma in synset.lemmas():
          syn.append(lemma.name())
    return syn

def head_word_synonyms_tokenizer(text):
    head_word_synonyms = []
    for head_word in head_word_tokenizer(text):
        head_word_synonyms.extend(get_synonyms(head_word))
    return head_word_synonyms
```

```python
head_words_synonyms_vectorizer = CountVectorizer(tokenizer = head_word_synonyms_tokenizer,max_features=100,stop_words=stopwords.words('english'))
head_words_synonyms_vector = head_words_synonyms_vectorizer.fit_transform(preprocessed_df["questions"].values).toarray()
```

### 5. Named Entities
Named entities are generated as bag of named enities with CountVectorizer function with the use of custom named entity tokenizer. Here original unprocessed questions are used since after preprocessing it may loose certain words.

```python
# https://stackoverflow.com/questions/24398536/named-entity-recognition-with-regular-expression-nltk
def named_entities_tokenizer(text):
    chunked = ne_chunk(pos_tag(word_tokenize(text)))
    prev = None
    continuous_chunk = []
    current_chunk = []

    for i in chunked:
        if type(i) == Tree:
            current_chunk.append(" ".join([token for token, pos in i.leaves()]))
        elif current_chunk:
            named_entity = " ".join(current_chunk)
            if named_entity not in continuous_chunk:
                continuous_chunk.append(named_entity)
                current_chunk = []
        else:
            continue
    if current_chunk:
        named_entity = " ".join(current_chunk)
        if named_entity not in continuous_chunk:
            continuous_chunk.append(named_entity)

    return continuous_chunk
```

```python
named_entities_vectorizer = CountVectorizer(tokenizer = named_entities_tokenizer,max_features=100,stop_words=stopwords.words('english'))
named_entities_vector = named_entities_vectorizer.fit_transform(preprocessed_df["questions"].values).toarray()
```

#### Confusion Matrix Plotter
Adapated from https://stackoverflow.com/questions/19233771/sklearn-plot-confusion-matrix-with-labels

```python
def cm_analysis(y_true, y_pred, labels, ymap=None, figsize=(8,8)):
    """
    Generate matrix plot of confusion matrix with pretty annotations.
    The plot image is saved to disk.
    args: 
      y_true:    true label of the data, with shape (nsamples,)
      y_pred:    prediction of the data, with shape (nsamples,)
      filename:  filename of figure file to save
      labels:    string array, name the order of class labels in the confusion matrix.
                 use `clf.classes_` if using scikit-learn models.
                 with shape (nclass,).
      ymap:      dict: any -> string, length == nclass.
                 if not None, map the labels & ys to more understandable strings.
                 Caution: original y_true, y_pred and labels must align.
      figsize:   the size of the figure plotted.
    """
    if ymap is not None:
        y_pred = [ymap[yi] for yi in y_pred]
        y_true = [ymap[yi] for yi in y_true]
        labels = [ymap[yi] for yi in labels]
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = ''
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
    cm = pd.DataFrame(cm, index=labels, columns=labels)
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(cm, annot=annot, fmt='', ax=ax)
    #plt.savefig(filename)
    plt.show()
```

### 2. A traditional ML classifier s.a. SVM or Logistic Regression with word embedding features
Logistic regression is used here and as earlier it is tested in two different ways. The word embeddings are loaded from word2vec-google-news-300 vectors.<br>
The embeddings for sentences are created in two different ways
    1. Mean embedding – Mean vector of words of a sentence
    2. TF-IDF mean embedding – Word of vectors are multiplied with TF-IDF scores and then retrieve the mean vector for a sentence 
https://stackoverflow.com/questions/29760935/how-to-get-vector-for-a-sentence-from-the-word2vec-of-tokens-in-sentence

```python
wv = api.load('word2vec-google-news-300')
```

#### 1. Mean Embeddings

```python
mean_embeddings = []
for sentence in preprocessed_df["preprocessed_questions"].values:
    sentence_vec = []
    for word in sentence.split():
        if word not in nltk.corpus.stopwords.words('english'):
            if word in wv.vocab.keys():
                sentence_vec.append(wv[word])
    if len(sentence_vec) == 0:
        sentence_vec = [[0]*300]
    mean_embeddings.append(np.array(sentence_vec).mean(axis=0))
```

```python
x_train, x_test, y_train, y_test = train_test_split(mean_embeddings, labels, test_size=0.2, random_state=42)
```

#### 2. TF-IDF Mean Embeddings

##### Dictionary creation and TF-DF score computation

```python
doc_list  = preprocessed_df["preprocessed_questions"].values
dictionary = Dictionary()
BoW_corpus = [dictionary.doc2bow(doc.split(), allow_update=True) for doc in doc_list]
tfidf = TfidfModel(BoW_corpus)   
```

##### Dictionary(data structure) is created for each word and its respective TF-IDF score

```python
tfidf_word_values = {}
tfidf_word_list = functools.reduce(operator.iconcat, tfidf[BoW_corpus], [])
for m in set(tfidf_word_list):
    tfidf_word_values[dictionary[m[0]]] = m[1]
```

```python
tfidf_mean_embeddings = []
for sentence in preprocessed_df["preprocessed_questions"].values:
    sentence_vec = []
    for word in sentence.split():
        if word not in nltk.corpus.stopwords.words('english'):
            if word in wv.vocab.keys():
                sentence_vec.append(np.array(wv[word]*tfidf_word_values[word]))
    if len(sentence_vec) == 0:
        sentence_vec = [[0]*300]
    tfidf_mean_embeddings.append(np.array(sentence_vec).mean(axis=0))
```

```python
x_train, x_test, y_train, y_test = train_test_split(tfidf_mean_embeddings, labels, test_size=0.2, random_state=42)
```

###### Logistic regression without cross validation

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)

y_pred = logisticRegr.predict(x_test)

print('accuracy %s' % accuracy_score(y_pred, y_test))
print(classification_report(y_test, y_pred,))
cm_analysis(label_encoder.inverse_transform(y_test), label_encoder.inverse_transform(y_pred), label_encoder.classes_)
```

###### Logistic regression with cross validation

```python
y_pred = cross_val_predict(logisticRegr, tfidf_mean_embeddings, labels, cv=10)

print('accuracy %s' % accuracy_score(y_pred, labels))
print(classification_report(labels, y_pred))
cm_analysis(label_encoder.inverse_transform(labels), label_encoder.inverse_transform(y_pred), label_encoder.classes_)
```

### 3. A NN classifier s.a. an LSTM for classification
As neural network classifier, LSTM is used. Bag of words is used as features and classes are transformed to one hot encoded vectors.

```python
max_fatures = 2000
tokenizer = Tokenizer(num_words=max_fatures, split=' ')
tokenizer.fit_on_texts(preprocessed_df["preprocessed_questions"].values)
x = tokenizer.texts_to_sequences(preprocessed_df["preprocessed_questions"].values)
x = pad_sequences(x)
y = to_categorical(labels)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Embedding(input_dim = max_fatures, output_dim = 128, input_length = x_train.shape[1]))
model.add(LSTM(128))
model.add(Dense(64, activation='relu'))
model.add(Dense(y_test.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
model.fit(x_train, y_train, epochs=5, batch_size=40, verbose = 2)
```

```python
y_pred = model.predict(x_test, batch_size=64, verbose=1)
y_pred_bool = np.argmax(y_pred, axis=1)
```

### 4. BONUS - experiment with a BERT-based classifier
Adapted from https://androidkt.com/simple-text-classification-using-bert-in-tensorflow-keras-2-0/

<div class="alert alert-info">

**Note:** Bert Classifier would take considerable time to run

```python
bert_layer=hub.KerasLayer("https://tfhub.dev/tensorflow/bert_en_uncased_L-12_H-768_A-12/1",trainable=True)

MAX_SEQ_LEN=128
input_word_ids = tf.keras.layers.Input(shape=(MAX_SEQ_LEN,), dtype=tf.int32,
                                       name="input_word_ids")
input_mask = tf.keras.layers.Input(shape=(MAX_SEQ_LEN,), dtype=tf.int32,
                                   name="input_mask")
segment_ids = tf.keras.layers.Input(shape=(MAX_SEQ_LEN,), dtype=tf.int32,
                                    name="segment_ids")

pooled_output, sequence_output = bert_layer([input_word_ids, input_mask, segment_ids])

FullTokenizer = bert_tokenization.FullTokenizer

vocab_file=bert_layer.resolved_object.vocab_file.asset_path.numpy()

do_lower_case=bert_layer.resolved_object.do_lower_case.numpy()

tokenizer=FullTokenizer(vocab_file,do_lower_case)

def get_masks(tokens, max_seq_length):
    return [1]*len(tokens) + [0] * (max_seq_length - len(tokens))

def get_segments(tokens, max_seq_length):
    """Segments: 0 for the first sequence, 1 for the second"""
    segments = []
    current_segment_id = 0
    for token in tokens:
        segments.append(current_segment_id)
        if token == "[SEP]":
            current_segment_id = 1
    return segments + [0] * (max_seq_length - len(tokens))

def get_ids(tokens, tokenizer, max_seq_length):
    """Token ids from Tokenizer vocab"""
    token_ids = tokenizer.convert_tokens_to_ids(tokens,)
    input_ids = token_ids + [0] * (max_seq_length-len(token_ids))
    return input_ids

def create_single_input(sentence,MAX_LEN):
  
  stokens = tokenizer.tokenize(sentence)
  
  stokens = stokens[:MAX_LEN]
  
  stokens = ["[CLS]"] + stokens + ["[SEP]"]
 
  ids = get_ids(stokens, tokenizer, MAX_SEQ_LEN)
  masks = get_masks(stokens, MAX_SEQ_LEN)
  segments = get_segments(stokens, MAX_SEQ_LEN)

  return ids,masks,segments

def create_input_array(sentences):

  input_ids, input_masks, input_segments = [], [], []

  for sentence in tqdm(sentences,position=0, leave=True):
  
    ids,masks,segments=create_single_input(sentence,MAX_SEQ_LEN-2)

    input_ids.append(ids)
    input_masks.append(masks)
    input_segments.append(segments)

  return [np.asarray(input_ids, dtype=np.int32), 
            np.asarray(input_masks, dtype=np.int32), 
            np.asarray(input_segments, dtype=np.int32)]
```

```python
train_sentences = preprocessed_df["preprocessed_questions"].values
list_classes = tags
y_binary = to_categorical(labels)
x_train, x_test, y_train, y_test = train_test_split(train_sentences, y_binary, test_size=0.2, random_state=42)
```

```python
x = tf.keras.layers.GlobalAveragePooling1D()(sequence_output)
x = tf.keras.layers.Dropout(0.2)(x)
out = tf.keras.layers.Dense(len(list_classes), activation="sigmoid", name="dense_output")(x)

model = tf.keras.models.Model(
      inputs=[input_word_ids, input_mask, segment_ids], outputs=out)

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
inputs=create_input_array(x_train)

model.fit(inputs,y_train,epochs=1,batch_size=32,validation_split=0.2,shuffle=True)
```

```python
test_inputs=create_input_array(x_test)

print(classification_report(np.argmax(y_test, axis = 1), np.argmax(model.predict(test_inputs), axis = 1)))
```
